```python
import pandas as pd
from transformers import pipeline

pd.options.display.max_colwidth = None
```

```python
# Initialize a sentiment analysis pipeline using a pre-trained XLM-RoBERTa model
clf = pipeline("sentiment-analysis", model="cardiffnlp/twitter-xlm-roberta-base-sentiment")
```

# Classification

```python
# Read the JSON file "tweets.json" into a pandas DataFrame
df = pd.read_json("tweets.json")
df
```

```python
# Identify parent tweets to prioritize them when dropping duplicates
df["is_parent"] = df["parent_tweet_id"] == df["tweet_id"]

# Sort by 'is_parent' to ensure parent tweets are processed last,
# then drop duplicates based on 'tweet_id', keeping the parent tweet if both parent and child versions exist.
df = df.sort_values("is_parent")
df = df.drop_duplicates(subset=["tweet_id"], keep="last")

df
```

## Predict Sentiment

```python
# Apply sentiment analysis to the 'tweet_text' column and extract the label
df["sentiment"] = clf(df["tweet_text"].to_numpy().tolist())
df["sentiment"] = df["sentiment"].apply(lambda value: value["label"])

df
```

## Sentiment Scoring

```python
# Count the occurrences of each unique username in the DataFrame
df.value_counts("username")
```

```python
# Define a dictionary to map sentiment categories to numerical scores
sentiment_score_map = {
    "negative": -1,
    "neutral" : 0,
    "positive": 1,
}

# Apply the mapping to create a new 'sentiment_score' column in the DataFrame
df["sentiment_score"] = df["sentiment"].map(sentiment_score_map)
df
```

```python
# Calculate the sum of sentiment scores for each user
df_user_sentiment_score = df.groupby("username").agg(
    user_sentiment_score = ("sentiment_score", "sum")
)

# Sort the DataFrame by user_sentiment_score
df_user_sentiment_score.sort_values("user_sentiment_score")
```

```python
# Merge sentiment scores with the main DataFrame
df = df.merge(df_user_sentiment_score, on="username")
df
```

```python
# Save the DataFrame to a JSON file with sentiment data
df.to_json("tweets_with_sentiment.json", orient="records", index=False, indent=4)
```
